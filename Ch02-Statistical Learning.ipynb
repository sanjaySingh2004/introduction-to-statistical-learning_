### *For Loops*

```python
# For loops is used when you have to do a repetitive task.
total = 0
for value in [3, 2,19]:
    total += value
print('Total is: {0}'.format(total))
```

```python
total = 0
for value in [3, 2,19]:
    for weight in [3,2,1]:
        total += value * weight
print('Total is: {0}'.format(total))
```

```python
total = 0
for value, weight in zip([2,3,19],
                         [0.2,0.3,0.5]):
    total += weight * value
print('Weighted average is: {0}'.format(total))
```

```python
import pandas as pd
```

```python
college = pd.read_csv("College.csv", index_col = 0)
college.head()
```

```python
 college.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix

cols = ['Top10perc', 'Apps', 'Enroll']

scatter_matrix(college[cols],
               figsize = (7,7), diagonal='hist')
plt.show()
```

```python
plt.figure(figsize = (5,5))
sns.boxplot(x = college['Private'], y = college['Outstate'])
plt.title('Outstate Tuition by Private/public Colleges')
plt.show()
```

```python
college['Elite'] = pd.cut(college['Top10perc'],
                          bins = [0,50,100],
                          labels = ['Yes', 'No'])
college['Elite'].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=college['Elite'], y=college['Outstate'])
plt.title('Outstate Tution for Elite vs Non-Elite Colleges')
plt.show()
```

```python
fig, axes = plt.subplots (2,2, figsize=(12,8))

# making a histogram of four variables
college['Apps'].hist(bins = 30, ax=axes[0,0])
axes[0,1].set_title('Applications Recieved')

college['Enroll'].hist(bins=30, ax=axes[0,1])
axes[0,1].set_title('New Enrollments')

college['Outstate'].hist(bins=30, ax=axes[1,0])
axes[1,0].set_title('Out-of-State Tution')

college['Grad.Rate'].hist(bins=30, ax=axes[1,1])
axes[1,0].set_title('Graduation Rate')

plt.tight_layout()
plt.show()
```

1. Private colleges generally have higher out-of-state tuition compared to public colleges.
2. Elite colleges (where more than 50% of students are in the top 10% of their high school class) tend to have significantly higher tuition.
3. The scatterplot matrix shows that the number of applications (Apps) and enrollments (Enroll) have a positive correlation.
4. Graduation rates mostly range between 50% and 100%, with a few exceptions.
5. The histograms indicate that applications, enrollments, and out-of-state tuition are right-skewed (i.e., most colleges have lower values, with a few having extremely high values).
